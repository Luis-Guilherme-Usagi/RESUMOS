### Física - Forças em trajetórias curvilíneas

$$ Força\:Centrípeta = \frac {massa\: \cdot\: velocidade² } {raio}$$

* $Normal= Resultante\:Centrípeta + Peso$, quando o móvel não está em um ponto alto da trajetória.


* $Normal = Peso - Resultante\:Centrípeta$, quando o móvel está em um ponto alto da trajetória.


$$Força\:Tangencial = Massa\:\cdot\: Aceleração\:Tangencial$$

$$Força\:Tangencial² = Força\:Resultante² - Força\:Centrípeta²$$

$$Força\:Tangencial = Massa\:\cdot\: Aceleração\:Tangencial$$

### Trabalho e Energia

* $Trabalho = Força \:\cdot\: Distância\:\cdot\:Cosseno \theta$, mas em alguns casos cosseno Θ assume valores que alteram a fórmula em algum aspecto final.


* Força aplicada na mesma direção e sentido do deslocamento. Nesse caso, o ângulo entre os vetores força e deslocamento é de 0º, sendo cos0º = 1, a fórmula se mantém.


* Força aplicada na mesma direção, contudo, em sentido oposto ao deslocamento. Nesse caso, o ângulo entre os vetores força e deslocamento é de 180º, sendo cos180º = -1, a fórmula torna-se: $Trabalho = -Força \:\cdot\: distância$


* Força aplicada perpendicularmente ao deslocamento, ou seja, executando um ângulo de 90º entre os vetores força e deslocamento. Nesse caso, o trabalho é nulo devido ao fato de cos90º = 0.

### Quanto a força peso e o trabalho realizado por ela:

* Na descida do corpo, o $Trabalho = Massa\:\cdot\:Gravidade\:\cdot\:Altura$


* Na subida do corpo, o $Trabalho = - Massa\:\cdot\:Gravidade\:\cdot\:Altura$


* Como dito anteriormente, caso os vetores força e deslocamento sejam perpendiculares, o $Trabalho = 0$


* O trabalho de uma força variável pode ser considerado numericamente equivalente ao somatório das áreas entre o gráfico e o eixo x, sendo a área abaixo do eixo horizontal representante do trabalho negativo, e acima, positivo.


$$Trabalho\:da\:Força\:Elástica = \frac{Constante\:elástica\:\cdot\: distância²}{2}$$. 


$$Potência = \frac{Trabalho}{ \Delta Tempo}$$


$$Energia\:mecânica = Energia\:Cinética + Energia\:Potencial$$


$$Energia\:Cinética = \frac{Massa\:\cdot\: Velocidade²}{2}$$


$$Energia\:Potencial = Massa \: \cdot\:Gravidade \:\cdot \:Altura$$

### Impulso e quantidade de movimento

$$Impulso= Força  \: \cdot\ \Delta Tempo$$


Quantidade de Movimento Inicial = Quantidade de Movimento Final ( Quando se analisa um sistema isento de forças externas)

### Classificação das colisões 

**Colisões Perfeitamente Elásticas**


* Coeficiente de Restituição = 1

* Quantidade de Movimento Inicial = Quantidade de Movimento Final

* Energia Cinética Inicial = Energia Cinética Final


**Colisões Perfeitamente Inelásticas**


* Coeficiente de Restituição = 0

* Quantidade de Movimento Inicial = Quantidade de Movimento Final

* Energia Cinética Inicial > Energia Cinética Final.
